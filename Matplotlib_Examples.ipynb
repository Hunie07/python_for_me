# â–¶ Decay Sine Response

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 100, 1000) 
tau = 10 # tau ê°€ ì»¤ì§ˆìˆ˜ë¡ ì§„í­ì´ ì¤„ì–´ë“œëŠ” ì •ë„ê°€ ìž‘ì•„ì§„ë‹¤.
y = np.sin(t) * np.exp(-t/tau)
plt.plot(t, y, label='Decay Oscillating Response')
plt.ylabel('y [m]')
plt.xlabel('x [s]')
plt.legend()
```

# â–¶ Euler Equation

```python
t = np.linspace(0, 1, 100)
f = 1                            # frequency (ì£¼íŒŒìˆ˜) [Hz]
y_euler = np.exp(1j*2*np.pi*f*t) # 1j : í—ˆìˆ˜
y_cos = np.real(y_euler)         # ì‹¤ìˆ˜ë¶€ë¶„
y_sin = np.imag(y_euler)         # í—ˆìˆ˜ë¶€ë¶„

hfig, hax = plt.subplots()
hax.plot(t, y_cos, '-r', label='cos')
hax.plot(t, y_sin, '--b', label='sin')
hax.grid()
hax.legend()
```

```python
# ê·¸ëž˜í”„ ì°½ ë‚˜ëˆ„ê¸°
hfig, hax = plt.subplots(2, 1) # 2í–‰ 1ì—´
hax[0].plot(t, y_cos, '-r', label='cos')
hax[1].plot(t, y_sin, '--b', label='sin')
hax[0].grid()
hax[0].legend()
hax[1].grid()
hax[1].legend()
```

# â–¶ 3D Plot

```python
x = np.linspace(0, 2*np.pi, 20)
y = np.linspace(0, 2*np.pi, 20)
grid_x, grid_y = np.meshgrid(x, y) # meshgrid : 1ì°¨ì› array ë¡œ 2ì°¨ì› grid ë¥¼ ìƒì„±

z = np.sin(grid_x) * np.sin(grid_y)

hfig = plt.figure()
hax = hfig.add_subplot(projection='3d')
hax.plot_surface(grid_x, grid_y, z, cmap='jet') # cmap='jet' : ë¹¨ê°•ë¶€í„° íŒŒëž‘ìƒ‰ê¹Œì§€
hax.set_xlabel('x')
hax.set_ylabel('y')
hax.set_zlabel('z')
```
