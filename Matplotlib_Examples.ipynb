# ▶ Decay Sine Response

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 100, 1000) 
tau = 10 # tau 가 커질수록 진폭이 줄어드는 정도가 작아진다.
y = np.sin(t) * np.exp(-t/tau)
plt.plot(t, y, label='Decay Oscillating Response')
plt.ylabel('y [m]')
plt.xlabel('x [s]')
plt.legend()
```

# ▶ Euler Equation

```python
t = np.linspace(0, 1, 100)
f = 1                            # frequency (주파수) [Hz]
y_euler = np.exp(1j*2*np.pi*f*t) # 1j : 허수
y_cos = np.real(y_euler)         # 실수부분
y_sin = np.imag(y_euler)         # 허수부분

hfig, hax = plt.subplots()
hax.plot(t, y_cos, '-r', label='cos')
hax.plot(t, y_sin, '--b', label='sin')
hax.grid()
hax.legend()
```

```python
# 그래프 창 나누기
hfig, hax = plt.subplots(2, 1) # 2행 1열
hax[0].plot(t, y_cos, '-r', label='cos')
hax[1].plot(t, y_sin, '--b', label='sin')
hax[0].grid()
hax[0].legend()
hax[1].grid()
hax[1].legend()
```

```python
data = np.random.randn(10000)
plt.hist(data, 100, density=True) # data 를 100개의 막대로 확률밀도함수

x = np.linspace(-4, 4, 100)
sigma = 1
mean = 0
nd = (  (1/(sigma*np.sqrt(2*np.pi)))*
    np.exp(-0.5*((x-mean)/sigma)**2)  )

plt.plot(x, nd, 'r', label='Std Normal Distribution')
plt.ylabel('PSD')
plt.xlabel('X')
plt.legend()
```

# ▶ 3D Plot

```python
x = np.linspace(0, 2*np.pi, 20)
y = np.linspace(0, 2*np.pi, 20)
grid_x, grid_y = np.meshgrid(x, y) # meshgrid : 1차원 array 로 2차원 grid 를 생성

z = np.sin(grid_x) * np.sin(grid_y)

hfig = plt.figure()
hax = hfig.add_subplot(projection='3d')
hax.plot_surface(grid_x, grid_y, z, cmap='jet') # cmap='jet' : 빨강부터 파랑색까지
hax.set_xlabel('x')
hax.set_ylabel('y')
hax.set_zlabel('z')
```

# ▶ Simple Animation

```python
fig, ax = plt.subplots()

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 1, 1, 1, 1])

hp, = ax.plot(x, y)
ax.set_ybound([0, 11]) # y축 범위 설정

for i in range(0, 11, 1):
    hp.set_ydata(i*y)
    plt.pause(0.5) # 0.5초마다 그래프 변화
# Animation 은 Spyder 에서 실행하자.
```
