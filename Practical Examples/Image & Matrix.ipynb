```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
# image size variables
height = 4
width = 5
depth = 3
```

```python
# black
M = np.zeros((height, width, depth))
plt.imshow(M)
plt.grid()
```

```python
# red
M[:, :, 0] = 255
plt.imshow(M)
plt.grid()
```

```python
# green
M = np.zeros((height, width, depth))
M[:, :, 1] = 255
plt.imshow(M)
plt.grid()
```

```python
# blue
M = np.zeros((height, width, depth))
M[:, :, 2] = 255
plt.imshow(M)
plt.grid()
```

```python
# white
M = np.ones((height, width, depth)) * 255
plt.imshow(M)
plt.grid()
```

```python
# black ver2
M = np.ones((height, width, depth)) * 0
plt.imshow(M)
plt.grid()
```

```python
# change the top-left pixel to white
M[0, 0, :] = 255
plt.imshow(M)
plt.grid(False)
```

```python
# change the top-left pixel to green
M[0, 0, 0] = 0
M[0, 0, 1] = 255
M[0, 0, 2] = 0
plt.imshow(M)
```

```python
# change the top-left pixel to blue
M[0, 0, 0] = 0
M[0, 0, 1] = 0
M[0, 0, 2] = 255
plt.imshow(M)
```

```python
# change the top-left pixel to red
M[0, 0, 0] = 255
M[0, 0, 1] = 0
M[0, 0, 2] = 0
plt.imshow(M)
```

```python
# back to black
M[:, :, :] = 0
plt.imshow(M)
```

```python
# change the first column to green
M[:, 0, 1] = 255
plt.imshow(M)
```

```python
# change the second column from the last to blue
M[:, -2, 2] = 255
plt.imshow(M)
```

```python
# change the third row to yellow
M[2, :, 0] = 255
M[2, :, 1] = 255
M[2, :, 2] = 0
plt.imshow(M)
```

```python
# extend in vertical direction
M_vt = np.vstack([M, M])
plt.imshow(M_vt)
```

```python
# extend in horizontal direction
M_hz = np.hstack([M, M])
plt.imshow(M_hz)
```

```python
# make it a little darker
M_hz = (M_hz/255) * 0.5
plt.imshow(M_hz)
```

```python
# checkerboard image
M = np.zeros((5, 5, 3))
plt.imshow(M)
```

```python
# turn into white
M[:, :, :] = 1 # 0부터 1의 숫자로 들어가도 된다.
plt.imshow(M)
```

```python
# turn into black
M[:] = 0
plt.imshow(M)
```

```python
# checkerboard
M[0::2, 0::2, :] = 1
plt.imshow(M)
```

```python
M[1::2, 1::2, :] = 1
plt.imshow(M)
```

```python
# Modifying Images using Logical Operations
N = M.copy()

# change white to gray
idx = np.where(N==1)
N[idx] = 0.5
plt.imshow(N)
```

```python
# change black to white
idx = np.where(N<0.1)
N[idx] = 1
plt.imshow(N)
```

```python
# copy again
N = M.copy()
plt.imshow(N)
```

```python
# color inverse
idx_w = np.where(N==1)
idx_b = np.where(N==0)
N[idx_w] = 0
N[idx_b] = 1
plt.imshow(N)
```

```python
# change white to green
idx1 = np.where(N==1)
N[idx1[0], idx1[1], 0] = 0
N[idx1[0], idx1[1], 2] = 0
plt.imshow(N)
# white 부분이라 이미 green 부분은 1 로 설정되어있다.
```

```python
# change black to red
flag_r = (N[:, :, 0]==0)
flag_g = (N[:, :, 1]==0)
flag_b = (N[:, :, 2]==0)

flag_blk = flag_r & flag_g & flag_b
idx_blk = np.where(flag_blk==True)

N[idx_blk[0], idx_blk[1], 0] = 1
plt.imshow(N)
```

```python
# change black to red (another way)
N = M.copy()
plt.imshow(N)
```

```python
N_sum = np.sum(N, axis=2) # RGB 숫자들의 합
idx_blk = np.where(N_sum==0) # black 이면 다 0 이니 합도 0이다.
N[idx_blk[0], idx_blk[1], 0] = 1
plt.imshow(N)
```

```python
# change diagonal to green
np.fill_diagonal(N[:, :, 0], 0)
np.fill_diagonal(N[:, :, 1], 1)
np.fill_diagonal(N[:, :, 2], 0)
plt.imshow(N)
```

```python
# jet scale
M = np.linspace(0, 1, 10)
M = M[:, np.newaxis]
M = np.repeat(M, 10, axis=1)
plt.imshow(M, cmap='gray')
```

```python
# transpose
M_tr = np.transpose(M)
plt.imshow(M_tr, cmap='gray')
```

```python
# transpose
M_tr = np.transpose(M)
plt.imshow(M_tr, cmap='jet')
```
