```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

▶ Example Plot 1 - Pyplot API Style

```python
y = np.array([10, 20, 15])
plt.plot(y)
```

```python
x = np.array([5., 10., 15.])
y = np.array([10., 20., 15.])
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r') # 'r' : 빨간색 선 / 'b' : 파란색 선
```

```python
plt.plot(x, y, 'b--') # 파란색 선 + 점선 (dashed style line)
```

```python
plt.plot(x, y, 'bo') # 파란색 원형 점
```

```python
plt.plot(x, y, 'b-o', label='Data 1') # 파란색 선 + 원형 점
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.legend() # label 표시 (범례)
```

▶ Example Plot 2 - Object-Oriented Style

```python
x = np.array([5., 10., 15.])
y = np.array([10., 20., 15.])
```

```python
hf = plt.figure()
ha = hf.add_axes([0, 0, 1, 1])
ha.plot(x, y)
```

```python
fig, ax = plt.subplots()
line, = ax.plot(x, y)
```

```python
x_new = np.array([10, 11, 12])
y_new = np.array([10, 10, 10])
line.set_data(x_new, y_new)
fig
```
