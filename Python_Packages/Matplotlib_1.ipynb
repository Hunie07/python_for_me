â–¶ Example Plot 2 - Object-Oriented Style

```python
x = np.array([5., 10., 15.])
y = np.array([10., 20., 15.])
```

```python
hf = plt.figure()
ha = hf.add_axes([0, 0, 1, 1])
ha.plot(x, y)
```

```python
fig, ax = plt.subplots()
line, = ax.plot(x, y)
```

```python
x_new = np.array([10, 11, 12])
y_new = np.array([10, 10, 10])
line.set_data(x_new, y_new)
fig
```
