# Sympy

```python
# symbols()
from sympy import symbols
x = symbols('x')
```

```python
# subs()
expr = 2 * x
expr.subs(x, 2) # x 에 2를 대입
```

```python
# differentiation (미분)
# diff()
f = x**3

from sympy import diff
df1 = diff(f, x) # 함수 f 를 x 에 대하여 미분
df2 = diff(df1, x)
df3 = diff(df2, x)
df1, df2, df3
```

```python
from sympy import sin
f = sin(x)

df1 = diff(f, x)
df1
```

```python
# integrate (적분)
# integrate()
from sympy import integrate
integrate(f, (x, 0, 2*3.14)) # 함수 f 를 x 에 대하여 0부터 2*3.14 까지 적분
```

```python
# limit (극한)
# limit()
from sympy import limit
limit(sin(x)/x, x, 0) # 해당 함수에서 x 가 0을 향해 갈때의 극한값
```

▶ 상미분방정식 (Ordinary Differential Equation)

```python
# ODE
from sympy import Function, Eq, dsolve
```

```python
t = symbols('t')
x = Function('x')
ODE = Eq(x(t).diff(t,t) + x(t), 0) # Eq(좌변, 우변)

dsolve(ODE, x(t)) # dsolve(ODE, answer to get) : ODE 해결 함수
```

# Scipy

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1., 0.8, 0.4, 0.3, 0.2])
```

```python
# plot
plt.plot(x, y, '*')
```

```python
# interpolate
f_lin = interpolate.interp1d(x, y) # 충족하는 1차함수 생성
x_new = np.arange(1, 5, 0.1)
y_lin = f_lin(x_new)
```

```python
# spline interpolation (곡선)
tck = interpolate.splrep(x, y, s=0)
y_spl = interpolate.splev(x_new, tck, der=0)
```

```python
# plot
fig, ax = plt.subplots()
ax.plot(x, y, 'o', label='Data')
ax.plot(x_new, y_lin, label='Linear')
ax.plot(x_new, y_spl, label='Spline')
ax.legend()
```
